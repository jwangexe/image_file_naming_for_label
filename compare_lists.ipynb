```python
ids = df.证书编号
arr = ids.values
print(arr, len(arr))

"H-RUKE18998" in arr
```

```python
import os
```

```python
print(os.listdir("./回单照片"))
print(len(os.listdir("./回单照片")))
print(len(set(os.listdir("./回单照片"))))
```

```python
# 表格上没有，压缩包里有
missing = []
for fname in os.listdir("./回单照片"):
    if fname.rstrip(".png") not in ids.values:
        print(fname)
        missing.append(fname.rstrip(".png"))
missing
```

```python
# 表格上有，压缩包里没有
missing_other = []
dir = os.listdir("./回单照片")
for fname in ids.values:
    if fname + ".png" not in dir:
        missing_other.append(fname)
missing_other[:5]
```

```python
len(missing_other)
```

```python
fhand = open("表格上有+压缩包里没有.txt", "w")
for fname in missing_other:
    print(fname, file=fhand)
fhand.close()
```

```python
# change cert number to ASPIRE number in separate folder
# 只改压缩包里有的
import shutil
ASPIRE_DIR = "./回单照片aspire"
IMAGE_DIR = "./回单照片"

for fname in os.listdir(IMAGE_DIR):
    idnum = fname.rstrip(".png")
    if idnum in missing:
        continue
    row = df.loc[df["证书编号"] == idnum]
    aspire = row["Aspire ID"].values[0]

    print(os.path.join(IMAGE_DIR, fname), os.path.join(ASPIRE_DIR, aspire+idnum+".png"))
    shutil.copy(os.path.join(IMAGE_DIR, fname), os.path.join(ASPIRE_DIR, aspire+idnum+".png"))
    
```
